```python
import json
import numpy as np
import os
import pandas
```

```python
results = os.listdir("../results/")
```

```python
results
```

```python
data = {}
for res in results:
  data[res] = []
```

```python
data
```

```python
for file in results:
  for percent in range(1, 17):
    file_name = f"../results/{file}/results_{percent}%.json"
    f = open(file_name, 'r')
    data[file].append({percent: json.loads(f.read())})
    f.close()
```

```python
data
```

```python
table_percentage = {"tpr": np.zeros((16, 3)), "fpr": np.zeros((16, 3)), "tnr": np.zeros((16, 3)), "fnr": np.zeros((16, 3))}
```

```python
for i, file in enumerate(data):
  print("going to ", file)
  for j in range(len(data[file])):
     experiment = data[file][j][j]
     for metric in experiment["test"]["ratios"]:
      # print(experiment["test"]["ratios"][metric])
      me = experiment["test"]["ratios"][metric]
      print(i, me)
      table_percentage[metric][j][i] = me
```

```python
table_percentage
```

```python
my_df = pandas.DataFrame(table_percentage)
my_df.index = np.arange(1, len(my_df) + 1)
```

```python
my_df
```

```python
my_df.to_latex()
```

```python
table_experiments = {"tpr": np.zeros((3, 16)), "fpr": np.zeros((3, 16)), "tnr": np.zeros((3, 16)), "fnr": np.zeros((3, 16))}
```

```python
for i, file in enumerate(data):
  for j in range(len(data[file])):
     experiment = data[file][j][j]
     for metric in experiment["test"]["ratios"]:
      table_experiments[metric][i][j] = experiment["test"]["ratios"][metric]
```

```python
for metric in table_experiments:
  table_experiments[metric] = np.mean(table_experiments[metric], axis=1)
```

```python
df_2 = pandas.DataFrame(table_experiments)
df_2.index = [exp[17:] for exp in data]
```

```python
df_2
```

```python
df_2.to_latex()
```

```python
import seaborn
import matplotlib.pyplot as plt
```

```python
myPlot = {}
for res in results:
  myPlot[res] = np.zeros((16))
```

```python
myPlot
```

```python
data
```

```python
for i in data:
  for ix, exp in enumerate(data[i]):
    x = list(exp.values())[0]
    cf = x["test"]["confusion_matrix"]
    myPlot[i][ix] = ((cf["tp"] + cf["tn"]) / (cf["tp"] + cf["fp"] + cf["fn"] + cf["tn"]))
```

```python
myPlot
```

```python
seaborn.set_palette('Set2')
seaborn.set_theme()
fig = plt.figure(figsize=(10,6))

seaborn.lineplot(myPlot["lenet_100_epochs_normalized probability_loss"])
seaborn.lineplot(myPlot["lenet_100_epochs_cross entropy_loss"])
seaborn.lineplot(myPlot["lenet_100_epochs_probability_loss"])

fig.legend(labels=['normalized probability loss','entr','probability loss'])

plt.title("FPR with different training data percentages")
plt.xlabel("Training Data Access %")
plt.ylabel("FPR")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

dx =[x for x in range(1, 17)]

seaborn.lineplot(x=dx, y=myPlot["lenet_100_epochs_normalized probability_loss"], data=myPlot, label='normalized probability loss')
seaborn.lineplot(x=dx, y=myPlot["lenet_100_epochs_cross entropy_loss"], data=myPlot, label='cross entropy loss')
seaborn.lineplot(x=dx, y=myPlot["lenet_100_epochs_probability_loss"], data=myPlot, label='probability loss')

ax.legend()

plt.title("Accuracy with different training data percentages")
plt.xlabel("Training Data Access %")
plt.ylabel("Accuracy")
plt.show()
```
